```python
def execution(sample):
    sample = sample.split('\n')
    procedures = [procedure.split(' ') for procedure in sample]
    values = []
    
    X = 1
    cycle = 1
    for procedure in procedures:
        if procedure[0] == 'noop':
            cycle += 1
            values = check_value(X, cycle, values)
            
        if procedure[0] == 'addx':
            X, cycle = cycle_addx(X, cycle, 0)
            values = check_value(X, cycle, values)
            X, cycle = cycle_addx(X, cycle, procedure[1])
            values = check_value(X, cycle, values)

    return sum(values)
```

```python
def cycle_addx(X, cycle, change):
    return X+int(change), cycle+1
```

```python
def check_value(X, cycle, values):
    if cycle in [20, 60, 100, 140, 180, 220]:
        values.append(X*cycle)
        
    return values
```
